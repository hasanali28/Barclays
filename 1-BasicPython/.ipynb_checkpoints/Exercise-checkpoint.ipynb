Q1. Write a function to detect wether a number is prime or not ? 

```python
def is_prime(num):
    """
    checks wether a number is prime or not
    """
    for i in range(2,num//2 + 1):
        if num % i == 0:
            return False
    
    return True
    
```

```python
for i in range(11):
    print (f"Number {i} is ", is_prime(i))
```
