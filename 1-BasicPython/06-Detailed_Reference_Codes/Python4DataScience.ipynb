***Nested Loop***

***Problem :*** Count vowels in a list of words.

```python
word_list = ['analytics','DC','standard','south africa','india','data']
vowel_list = ['a','e','i','o','u']
totalCount = 0
for word in word_list :
    # initialize variable count with value 0
    count = 0
    
    # now iterate over characters of each word
    for character in word :
        
        # if the character is found in the list then increase the value of count variable
        if(character in vowel_list):
            count = count+1
    # print the word and vowel count of that word    
    print(word,count)
    
    totalCount = totalCount + count

print('The total vowel count in the word list is ', totalCount)
```

```python
# printing the same using the while loop
word_list = ['analytics','dc','standard','south africa','india','data science']
vowel_list = ['a','e','i','o','u']
totalCount = 0
i = 0
while(i<len(word_list))  :
    # initialize variable count with value 0
    count = 0
    
    # now iterate over characters of each word
    j = 0
    while(j<len(word_list[i])) :
        # if the character is found in the list then increase the value of count variable
        if(word_list[i][j] in vowel_list):
            count = count+1
        j += 1
      
    # print the word and vowel count of that word    
    print(word_list[i],count)
    i += 1
    totalCount = totalCount + count

print('The total vowel count in the word list is ', totalCount)
```

***REDUCE FUNCTION***

***The reduce() function in Python takes in a function and a list as argument. The function is called with a lambda function and a list and a new reduced result is returned. This performs a repetitive operation over the pairs of the list. ***

---



```python
# nested_list = [[1,2,3],[2,3,5],[5,7,8]]

from functools import reduce
original_list = [1, 2, 3, 4, 5, 6]
org_list_str = ['asd', 'nbbn', 'kjkj']
# use reduce function to find the sum of numbers in the original list
sum_of_list = reduce((lambda x , y : x + " " + y), org_list_str)
print ('Sum of Numbers in the original list is : ',sum_of_list) 
```

```python
original_list = [110, 53, 3, 424, 255, 16, 42, 756]  
# use reduce function to find the largest number in the original list
largest_number = reduce((lambda x, y: x if (x > y) else y ), original_list) 
print ('Largest Number in the original list is : ',largest_number) 
```

***UNION, INTERSECTION AND DIFFERENCE OF SETS***

```python
# UNION OF SETS
# A U B = A + B - (A intersceton B)

setA = { 1, 2, 3, 4, 5 }
setB = { 4, 5, 6, 7, 8 }

setC = setA.union(setB)
print(setC)
```

```python
# INTERSECTION OF SETS

setC = setA.intersection(setB)
print(setC)
```

```python
# DIFFERENCE OF SETS

print(setA)
print(setB)
setC = setA.difference(setB)
print(setC)

setD = setB.difference(setA)
print(setD)

setX = setA - setB
print(setX)

setY = setB - setA
print(setY)


setSymmDiff = setA.symmetric_difference(setB)
setSymmDiff
```
